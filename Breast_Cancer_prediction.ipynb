```python
# import libraries
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Importing data and create dataframe

```python
# load dataset from csv file
df=pd.read_csv('/content/data.csv')
print(df)
```

```python
df.dtypes
```

```python
df.head()
```

Diagnosis column type is object.
M=Malignant, B=Benign
Change datatype to int(boolean, 0s and 1s)


```python
#use dummy variable to create new columnn
dum=pd.get_dummies(df['diagnosis'])
dum.rename(columns={'M':'Malignant','B':"Benign"},inplace=True)

#concatenate the new variables, and drop the previous variable 
df2=pd.concat([df,dum],axis=1)
df2.drop("diagnosis",axis=1,inplace=True)
```

```python
df2.tail()
```

```python
# Now only use Malignant columnt as new diagnosis columns
df2.drop(['Benign'],axis=1,inplace=True)
```

```python
df2.tail()
```

```python
# rename "Malignant" as 'Diagnosis'
df2.rename(columns={'Malignant':"Diagnosis"},inplace=True)
```

```python
# 'id' and 'Unnaed:32' are not useful data, drop
df2.drop(['id','Unnamed: 32'], axis=1, inplace=True)
```

```python
# change dtype into int64
df2['Diagnosis']= df2['Diagnosis'].astype('int')
```

Separate the data from target(diagnosis)

```python
X=df2.drop(columns='Diagnosis', axis=1)
Y=df2["Diagnosis"]
```

```python
print(X)
```

```python
print(Y)
```

Split data into training and test data

```python
# split data into 30% test data, 70% training
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y, random_state=0)
```

```python
# show division of data into training and test data
print(X.shape, X_train.shape, X_test.shape)
```

Train Model

```python
model=LogisticRegression(class_weight='balanced',max_iter=10000)
```

```python
# train model with training data
model.fit(X_train,Y_train)
```

Test Accuracy

```python
X_prediction=model.predict(X_train)
```

```python
score=accuracy_score(Y_train, X_prediction)
print(score)
```

```python
X_test_prediction=model.predict(X_test)
score_test=accuracy_score(Y_test, X_test_prediction)
print(score_test)
```

Build predictive test

```python
# features from individual with benign tumor
data=(7.76,24.54,47.92,181,0.05263,0.04362,0,0,0.1587,0.05884,0.3857,1.428,2.548,19.15,0.007189,0.00466,0,0,0.02676,0.002783,9.456,30.37,59.16,268.6,0.08996,0.06444,0,0,0.2871,0.07039)
data_numpyArray=np.asarray(data)
```

```python
#reshape numpy array (predict for one data point)
data_reshape = data_numpyArray.reshape(1,-1)
```

```python
#prediction
prediction=model.predict(data_reshape)
print(prediction)
```
